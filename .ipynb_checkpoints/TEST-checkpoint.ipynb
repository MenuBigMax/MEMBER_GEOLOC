```python
#------------------------ 
# USER DEFINED FUNCTIONS
#------------------------ 

from geopy.geocoders import Nominatim

# Return a Pandas series with 4 fields :
# - flg_geoloc_ok : Did Nominatim find geomapping ? (T/F)
# - address_words : Searched address in words
# - address_lat: latitude if finded, nan else
# - address_long: longitude if finded, nan else

def geoloc_or_log(adrs):
    geocoder = Nominatim(user_agent="moi_meme_compagny")
    try:
        gl=geocoder.geocode(adrs, True, 30)
        s=[True, adrs, gl.latitude, gl.longitude]
    except:
        s=[False, adrs, float('NaN'), float('NaN')]
    return(pd.Series(s, index=['flg_geoloc_ok', 'address_words', 'address_lat', 'address_long']))
```

```python
import pandas as pd

# Input file with addresses in human words
L=pd.read_csv("data/20181202_adherents_rscc.csv", sep=";")
# Searching geomapping for each address
G=(L['Adresse']+','+L['ville']+',FRANCE').apply(geoloc_or_log)
# Export of results (with no-response too)
G.to_csv('data/result.csv', sep=";", header=True, index_label='id_raw')
```
